```python
from transformers import T5ForConditionalGeneration, AutoTokenizer

mycheckpoint = "/home/kris/Project/Proposed-working/CodeCrafter/saved_models/checkpoint-120/"
device = "cpu" # for GPU usage or "cpu" for CPU usage

tokenizer = AutoTokenizer.from_pretrained(mycheckpoint)
model = T5ForConditionalGeneration.from_pretrained(mycheckpoint).to(device)

```

```python
query=input()
inputs = tokenizer.encode(query, return_tensors="pt").to(device)
outputs = model.generate(inputs, max_length=100)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
